```python

def dist(x1,x2,y1,y2,z1,z2):
    return np.sqrt((x1-x2)**2 + (y1-y2)**2 + (z1-z2)**2)
bins = np.linspace(0,200,40)
scent = (bins[1:] + bins[:-1])/2
dd_n = 0
dr_n = 0

jj = [0]
dj = 1000
for mm in tqdm(jj):
    ii = (mock1['x'] < mm +dj).__and__(mock1['x'] >= mm).__and__(mock1['y'] <mm+dj).__and__(mock1['y'] >=mm).__and__(mock1['z'] <mm+dj).__and__(mock1['z'] >=mm) 
    tt = mock1.loc[ii]
    x ,y ,z = tt['x'].to_numpy() , tt['y'].to_numpy() , tt['z'].to_numpy() 
    x_0 ,y_0 ,z_0 = tt['x'].to_numpy() , tt['y'].to_numpy() , tt['z'].to_numpy() 
    DD = []
    DR= []
    for i in tqdm(range(len(x_0))): # range(len(x_0))
        x_i = x_0[i]
        y_i = y_0[i]
        z_i = z_0[i]
        x = np.delete(x,0,0)
        y = np.delete(y,0,0)
        z = np.delete(z,0,0)
        dd = dist(x_i,x,y_i,y,z_i,z)
        kk = pd.cut(dd,bins)

        dd_n += kk.describe()


    x ,y ,z = tt['x'].to_numpy() , tt['y'].to_numpy() , tt['z'].to_numpy() 
    DR_x = stats.uniform(x.min(),x.max()).rvs(len(x)*3)
    DR_y = stats.uniform(y.min(),y.max()).rvs(len(y)*3)
    DR_z = stats.uniform(z.min(),z.max()).rvs(len(z)*3)

    DR_x0 = DR_x 
    DR_y0 = DR_y 
    DR_z0 = DR_z 

    for i in tqdm(range(len(DR_x0))):
        DR_xi = DR_x0[i]
        DR_yi = DR_y0[i]
        DR_zi = DR_z0[i]

        DR_x = np.delete(DR_x ,0,0)
        DR_y  = np.delete(DR_y,0,0)
        DR_z = np.delete(DR_z,0,0)

        dr = dist(DR_xi,DR_x,DR_yi,DR_y,DR_zi,DR_z)
        kk = pd.cut(dr,bins)
        dr_n += kk.describe()
```

```python
ff = (dd_n/dr_n)

nor = (len(x)*(len(x)-1))/(((len(x)*3)*(len(x)*3-1)))


xi = ff/nor -1
xi = xi['counts'].dropna()


bins = np.linspace(0,200,39)
plt.plot(bins,scent *scent *xi,'ro')

```
